## What is Machine Learning?

Machine Learning (ML) is a way of teaching computers to learn from data and make predictions or decisions without being explicitly programmed for every situation.

Instead of writing thousands of rules, you give the computer examples, and it learns the patterns by itself.

<br></br>
Example:

Imagine you want to teach a child to recognize cats.
You don't explain every possible cat feature. Instead, you show many pictures:

🐱 ---> Cat

🐱 ---> Cat

🐱 ---> Cat

🐶 ---> Not a cat

🐶 ---> Not a cat


After seeing enough examples, the child learns what a cat looks like.


<br></br>
Machine learning works similarly:

- Give the computer lots of data.

- The computer finds patterns in the data.

- It uses those patterns to make predictions on new data.
<br></br>

Real-World Examples:
<br></br>

> Email Spam Detection

- Learns from thousands of spam and non-spam emails.
- Predicts whether a new email is spam.

<br></br>

> Netflix or YouTube Recommendations

- Learns what you watch.
- Suggests similar content.

<br></br>

> Face Recognition

- Learns facial features.
- Identifies people in photos.

<br></br>

> House Price Prediction

- Learns from past house sales.
- Predicts the price of a new house.

<br></br>

Also, keep in mind, Machine Learning is a subfield of Artificial Intelligence.

### Traditional Programming vs Machine Learning

1. Traditional Programming

> Rules/Algorithms + Data → Answer/Results
- You write the rules.
<br></br>

Example:
```bash
if score >= 50:
    print("Pass")
else:
    print("Fail")
```

<br></br>
2. Machine Learning

> Data + Answers/Results → Rules/Algorithms (Model)
- The computer discovers the rules from the data.
